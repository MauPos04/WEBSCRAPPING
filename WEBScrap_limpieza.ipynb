```python
df_data_faltante['home'] = df_data_faltante['home'].str.strip()
df_data_faltante['away'] = df_data_faltante['away'].str.strip()
```

## Combinar faltante y historica

```python
if all(df_data_historical == df_data_faltante):
    print("Los dataframes son iguales")
else:
    print("Los dataframes no son iguales")
```

Como son iguales no los voy a concatenar


```python
df_data_historical2 = pd.concat([df_data_historical, df_data_faltante], ignore_index=True)
df_data_historical2.drop_duplicates(inplace=True)
df_data_historical2.sort_values('year', inplace=True)
df_data_historical2
```

Verifico que no esten duplicados

```python
df_brasil = df_data_historical2.query("year ==2002")
df_brasil.groupby("home")["year"].sum()
```

Haciendo ensayos para despues un posible xgboost

```python
copia1 = df_data_historical2.copy()
copia1["home"] = copia1["home"].astype(str)
copia1
labels, unique = pd.factorize(copia1['home'])

# Reemplazar la columna 'home' con las etiquetas
copia1['home'] = labels

copia1
```

Quitemos los guiones en los penales

```python
# df_data_historical2[df_data_historical2['score'].str.contains('[()]')]
# df_data_historical2['score'].str.replace('[()]')
# df_data_historical2['score'] = df_data_historical2['score'].str.replace('[()]', '', regex=True)
df_data_historical2['score'] = df_data_historical2['score'].str.replace('[()aet.]', '', regex=True)
```

```python
df_data_historical2['score'].str.contains('[^d–]')
```

```python
df_data_historical2['home'] = df_data_historical2['home'].str.strip()
df_data_historical2['away'] = df_data_historical2['away'].str.strip()
df_data_historical2['score'] = df_data_historical2['score'].str.strip()
```

```python
## ojo que ese guion no es un guion normal y aqui puse para que fuera independiente del guion
df_data_historical2[['HomeGoals', 'AwayGoals']] = df_data_historical2['score'].str.split('[-–]', expand=True)


```

```python
df_data_historical2.drop('score', axis=1, inplace=True)
```

```python
df_data_historical2.rename(columns={'home': 'HomeTeam', 'away' : 'Awayteam', 'year' : 'Year'}, inplace=True)
```

```python
df_data_historical2
```

```python
df_data_historical2.dtypes
df_data_historical2 = df_data_historical2.astype({'HomeGoals': int, 'AwayGoals' : int, 'Year': int} )
```

```python
df_data_historical2.dtypes
```

```python
df_data_historical2['TotalGoals'] = df_data_historical2['HomeGoals']+ df_data_historical2['AwayGoals']
```
